```python
# Scatter Plot Game
import matplotlib.pyplot as plt
import random

score = 0

h = int(input("Choose your difficulty level 1-3: "))

if h == 1:
  xmin = -10
  xmax = 10
  ymin = -10
  ymax = 10
if h == 2:
  xmin = -13
  xmax = 13
  ymin = -13
  ymax = 13
if h == 3:
  xmin = -16
  xmax = 16
  ymin = -16
  ymax = 16


fig, ax = plt.subplots()

for i in range(0,3):
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    x = [xpoint]
    y = [ypoint]
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'b') # blue x axis
    plt.plot([0,0],[ymin,ymax], 'b') # blue y axis
    plt.plot(x, y, 'ro')
    print(" ")
    plt.grid() # displays grid lines on graph
    plt.show()
    guess = input("Enter the coordinates of the red point point: \n")
    guess_array = guess.split(",")
    xguess = int(guess_array[0])
    yguess = int(guess_array[1])
    if xguess == xpoint and yguess == ypoint:
        score = score + 1

print("Your score: ", score) # notice this is not in the loop
```

```python
# Projectile Game
import numpy as np
import matplotlib.pyplot as plt
from sympy import*
%matplotlib inline
from ipywidgets import interactive
import random


wall_distance  = random.randint(-10, 10)
wall_height = random.randint(1, 8)


xmin = -10
xmax = 10
ymin = -10
ymax = 10

linex = [wall_distance, wall_distance]
liney = [0, wall_height]

points = 2*(xmax-xmin)
def f(a,b,c):
  x = np.linspace(xmin, xmax, points)
  y = a*x**2 + b*x + c

  fig, ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax,],[0,0], 'b')
  plt.plot([0,0],[ymin,ymax], 'b')
  plt.plot(linex, liney, 'r')

  plt.plot(x,y)
  plt.show()

interactive_plot = interactive(f,a=(-50,50),
b = (-50,50),c=(-50,50))

def manual_values():
  x = np.linspace(xmin,xmax,points)

  fig,ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax], [0,0], 'b')
  plt.plot([0,0], [ymin, ymax], 'b')
  plt.plot(linex, liney, 'r')

  plt.show()
  a = int(input("Enter value. a= "))
  b = int(input("Enter value. b= "))
  c = int(input("Enter value. c= "))
  y = a*x**2 + b*x + c
  fig, ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax], [0,0], 'b')
  plt.plot([0,0], [ymin, ymax], 'b')
  plt.plot(linex, liney, 'r')
  plt.plot(x,y)
  plt.show


z = int(input("choose difficulty level 1 or 2: "))
if z == 1:
  interactive_plot
elif z==2:
  manual_values()


# Run the code once, then change the code and run it again
```
